```python
accidents_severity = df1.groupby('Severity').count()['ID']
accidents_severity
```

```python
import matplotlib.pyplot as plt

# Data for accidents severity
accidents_severity = [1, 2, 3, 4]

# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the pie chart
wedges, texts, autotexts = ax.pie(accidents_severity, labels=accidents_severity, autopct='%1.1f%%', pctdistance=0.85)

# Set title
ax.set_title("Accident by Severity", fontdict={'fontsize': 16})

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Show the plot
plt.tight_layout()
plt.show()
```
